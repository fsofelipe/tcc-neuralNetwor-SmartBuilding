# MODELO d3:
Qualquer nível de melhora<br>
### Arquivos:
- Modelo aberto<br>
- Modelo fechado<br>
- 30 modelos random

```python
def is_better(dataFrame, it, PMV_label1, PMV_label2):
    """it = iterator"""
    it = int(it)
    if (it > 0 and it < (dataFrame['Hora'].count() - 1)):
        if ((abs(dataFrame.iloc[it - 1][PMV_label1]) >= abs(dataFrame.iloc[it][PMV_label1])) and
            (abs(dataFrame.iloc[it - 1][PMV_label2]) >= abs(dataFrame.iloc[it][PMV_label2]))):
            return True
        else:
            return False
    else:
        return False
    
    #return (dataFrame.iloc(it))
```

SELECIONA A PARTIR DE **TODAS ABERTAS** E **TODAS FECHADAS**

```python
selected = pd.DataFrame(columns=header)
tot = 0
c = 0

for i in xrange(1, all1['Mes'].count() - 1):
    if (is_better(all1, i, 'Sala_PMV', 'Mezanino_PMV')):
        #print i
        c = c + 1
        selected = selected.append(all1.iloc[i])

print "All_1: " +  str((c * 100.0)/8760.0) + "%" + " | total: " + str(c)


tot = tot + c
c = 0
for i in xrange(1, all0['Mes'].count() - 1):
    if (is_better(all0, i, 'Sala_PMV', 'Mezanino_PMV')):
        #print i
        c = c + 1
        selected = selected.append(all0.iloc[i])

print "All_0: " +  str((c * 100.0)/8760.0) + "%" + " | total: " + str(c)
```

# MODELO d5:
Melhora significativa 10% ou mais de conforto entre a hora anterior e o presente momento<br>
### Arquivos:
- Modelo aberto<br>
- Modelo fechado<br>
- 150 modelos random<br>
- Sistema Especialista

```python
def is_better_sig(dataFrame, it, PMV_label1, PMV_label2):
    """it = iterator"""
    it = int(it)
    if (it > 0 and it < (dataFrame['Hora'].count() - 1)):
        if (((abs(dataFrame.iloc[it - 1][PMV_label1]) - abs(dataFrame.iloc[it][PMV_label1])) >= (abs(dataFrame.iloc[it][PMV_label1]) * 0.1)) and
            ((abs(dataFrame.iloc[it - 1][PMV_label2]) - abs(dataFrame.iloc[it][PMV_label2])) >= (abs(dataFrame.iloc[it][PMV_label2]) * 0.1))):
            return True
        else:
            return False
    else:
        return False
```
